# IPython with bash

IPython can interact with other languages besides python. In the code below, a simple bash script is run to download and clean up UN data on apricot production.

```python
%%bash
curl > "../input/apricot_data.zip" "http://data.un.org/Handlers/DownloadHandler.ashx?DataFilter=itemCode:526&DataMartId=FAO&Format=csv&s=countryName:asc,elementCode:asc,year:desc&c=2,3,4,5,6,7&"
unzip -p ../input/apricot_data.zip > ../input/apricot_data.csv
sed -i -r 's/([a-z]),/\1/' ../input/apricot_data.csv  #remove delimiter from country names, i.e. China, mainland
grep '+' ../input/apricot_data.csv > ../input/apricot_regions.csv
grep -v '+' ../input/apricot_data.csv > ../input/apricot_countries.csv
```

# Using pandas to quickly inspect a data set

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

Read in the data:

```python
data = pd.read_csv('../input/apricot_countries.csv')
data.head()
```

Reshape the data to have year as the index, country as the columns, and area harvested as the value:

```python
ahdata = data[data['Element'] == 'Area Harvested']
ahdata = ahdata.pivot(index='Year', columns='Country or Area', values='Value')
ahdata.ix[:, :10].head()
```

Plot area harvested for each country:

```python
for col in ahdata.columns:
    plt.figure(figsize=[8,3])
    plt.title(col)
    plt.ylabel('Area Harvested (Ha)')
    years = [int(y) for y in ahdata.index.values]
    plt.plot(years, ahdata.ix[:, col])
    plt.xticks(years, years, rotation=90, size='small')
```
